```python
# Dependencies
import pandas as pd
import plotly.express as plt
import os
import statsmodels.api as sm
import numpy as np
```

```python
bikecounts_file = os.path.join('../','Resources', 'bikecounts.csv')
bikecounts_df = pd.read_csv(bikecounts_file)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])
bikecounts_df.head(6)
```

```python
#Hypothesis: A lower value for low temp will lead to a lower total bike count
```

```python
# Select your independent X variables, and your dependent y variable. Note: Be sure to add your constant to your X term.
X=bikecounts_df[["Low Temp (°F)"]]
Y=bikecounts_df[["Total"]]
X=sm.add_constant(X)
model=sm.OLS(Y,X)
results=model.fit()
```

```python
# Print the linear regression results
print(results.summary())
```

```python
#There is not ebough evidence to to reject the null hypothesis as the  R-squared is quite low
```
