```python
import numpy as np
import pandas as pd
import plotly.express as px
import scipy.stats as sts
```

```python
df = pd.read_csv('Resources/property_sales.csv')
df = df.select_dtypes(exclude=['object'])
df = df.drop('PrefixDire', 1)

sales=df.dropna()

sales
```

```python
for (columnName, columnData) in sales.iteritems():
    print(sts.normaltest(columnData.values))



```

```python
sample1 = sales.sample(10)
sample2 = sales.sample(10)
sts.ttest_ind(sample1, sample2, equal_var=False)
```

```python

#For the most part the t statistic is quite low which means the 2 samples were definintly derived from the same sample
```
