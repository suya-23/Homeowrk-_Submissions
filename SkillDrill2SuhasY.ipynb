```python
# Import Dependencies
import pandas as pd
import os 
```

```python
# Get the file path for running team one
file_path = os.path.join('../','Resources', 'team-1.csv')
```

```python
# Read the CSV file into a DataFrame
team1_scores_df = pd.read_csv(file_path)
team1_scores_df=team1_scores_df.drop(['coach_name'], axis=1)
team1_scores_df.head()
```

```python
# Melt the DataFrame
team1_scores_df.melt() 
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team1_scores_df.melt(id_vars="runner_name") 
```

```python
# Renaming the columns
team1_scores_df = team1_scores_df.melt(id_vars="runner_name", var_name="run", value_name="score")
team1_scores_df
```

```python
# Save results to a file
team1_scores_df.to_csv("results.csv", index=False, header=True) 
```
